```python
import pandas as pd
import numpy as np
import selenium
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
```

```python
chrome_path = r"C:\Users\Abhay Mahajan\Downloads\chromedriver.exe"
```

```python
driver = webdriver.Chrome(chrome_path)
```

```python
hrefs = []
for i in range(119):
    driver.get('https://seeclickfix.com/watchers/23686?page=' + str(i))
    issues = driver.find_elements_by_class_name('riverTitle')
    for issue in issues:
        elements = issue.find_elements_by_tag_name("a")
        hrefs.append(elements[0].get_attribute("href"))
```

```python
title = driver.find_element_by_xpath("//span[@itemprop = 'name']").text
location = driver.find_element_by_xpath("//div[@class = 'tagline']").text
description = driver.find_element_by_xpath("//p[@class = 'mts mbm']").text
```

```python
descriptions = []
for href in hrefs:
    driver.get(href)
    title = driver.find_element_by_xpath("//span[@itemprop = 'name']").text
    location = driver.find_element_by_xpath("//div[@class = 'tagline']").text
    description = driver.find_element_by_xpath("//p[@class = 'mts mbm']").text
    descriptions.append({
        'title' : title,
        'location' : location,
        'description' : description
    })
```

```python
date_reported = []
for href in hrefs:
    driver.get(href)
    date_posted = driver.find_element_by_xpath("//div[@class = 'txt-block']/time").text
    date_reported.append(date_posted)
```

```python
len(descriptions)
```

```python
df = pd.DataFrame(descriptions)
```

```python
df.insert(loc = 1 , column = 'date_reported' , value = date_reported)
```

```python
df.to_csv('SeeClickFix.csv')
```
