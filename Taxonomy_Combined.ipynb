```python
def bow(df):
    corpus = []
    for i in range(len(df)):
        review = df.loc[i , 0]
        re.sub('[^a-zA-Z]', ' ', review)
        review = review.lower()
        review = review.split()
        review = [wordnet.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
        review = ' '.join(review)
        corpus.append(review)
    return corpus
    
```

```python
corpus_accessibility = bow(df_accessibility)
corpus_amenities = bow(df_amenities)
corpus_architecture = bow(df_architecture)
corpus_cleanliness = bow(df_cleanliness)
corpus_construction = bow(df_construction)
corpus_cycling = bow(df_cycling)
corpus_parking = bow(df_parking)
corpus_maintenance = bow(df_maintenance)
corpus_green_spaces = bow(df_green_spaces)
corpus_sports_facilities = bow(df_sports_facilities)
corpus_walkability = bow(df_walkability)
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
stopwords += [',' , '(' , ')' , '.' , '/' , '!' , ';' , ':' , '-' , '_' , '?']
```

```python
def frequency(corpus):
    word_frequencies = {}
    corpus = ' '.join(corpus)
    for word in nltk.word_tokenize(corpus):
        if word not in stopwords:
            if word not in word_frequencies.keys():
                word_frequencies[word] = 1
            else:
                word_frequencies[word] += 1
    word_frequencies = sorted(word_frequencies.items() , key = lambda x : x[1] , reverse = True)
    return word_frequencies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
def taxonomy(corpus):
    X = cv.fit_transform(corpus).toarray()
    features = cv.get_feature_names()
    df_features = pd.DataFrame(X , columns = features)
    return features
```
